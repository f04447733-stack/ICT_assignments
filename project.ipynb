```python
books = {}
students = {}
issued = {}

MAX_BOOKS = 3
FINE_PER_DAY = 2

while True:
    print("\n--- SMART LIBRARY SYSTEM ---")
    print("1. Add Book")
    print("2. Register Student")
    print("3. Issue Book")
    print("4. Return Book")
    print("5. Search Book")
    print("6. Overdue Report")
    print("7. Exit")

    choice = input("Enter choice: ")

    # 1. Add Book
    if choice == "1":
        bid = input("Book ID: ")
        title = input("Title: ")
        author = input("Author: ")
        total = int(input("Total Copies: "))

        books[bid] = [title, author, total, total]
        print("Book added successfully")

    # 2. Register Student
    elif choice == "2":
        sid = input("Student ID: ")
        name = input("Student Name: ")

        students[sid] = name
        issued[sid] = []
        print("Student registered")

    # 3. Issue Book
    elif choice == "3":
        sid = input("Student ID: ")
        bid = input("Book ID: ")

        if sid in students and bid in books:
            if books[bid][3] > 0:
                if len(issued[sid]) < MAX_BOOKS:
                    already = False
                    for item in issued[sid]:
                        if item[0] == bid:
                            already = True

                    if already == False:
                        days = int(input("Enter days to borrow: "))
                        issued[sid].append([bid, days])
                        books[bid][3] -= 1
                        print("Book issued successfully")
                    else:
                        print("Same book already issued to this student")
                else:
                    print("Student reached maximum book limit")
            else:
                print("No copies available")
        else:
            print("Invalid student or book ID")

    # 4. Return Book
    elif choice == "4":
        sid = input("Student ID: ")
        bid = input("Book ID: ")

        if sid in issued:
            found = False
            for item in issued[sid]:
                if item[0] == bid:
                    days_taken = int(input("Enter days taken: "))
                    allowed = item[1]

                    if days_taken > allowed:
                        fine = (days_taken - allowed) * FINE_PER_DAY
                        print("Overdue Fine: ₹", fine)
                    else:
                        print("Book returned on time")

                    issued[sid].remove(item)
                    books[bid][3] += 1
                    found = True

            if found == False:
                print("Book not found in issued list")
        else:
            print("Invalid student")

    # 5. Search Book (Partial Match)
    elif choice == "5":
        keyword = input("Enter title keyword: ").lower()

        for bid in books:
            if keyword in books[bid][0].lower():
                print(bid, "-", books[bid][0], "-", books[bid][3], "copies available")

    # 6. Overdue Report
    elif choice == "6":
        print("\n--- OVERDUE REPORT ---")
        for sid in issued:
            for item in issued[sid]:
                if item[1] < 0:
                    print(students[sid], "has overdue book", item[0])

    # 7. Exit
    elif choice == "7":
        print("Exiting Library System")
        break

    else:
        print("Invalid choice")
```

```python
inventory = {}
LOW_STOCK_LIMIT = 5

while True:
    print("\n--- CAMPUS STORE SYSTEM ---")
    print("1. Add Product")
    print("2. Update Stock")
    print("3. Purchase Products")
    print("4. View Inventory")
    print("5. Exit")

    choice = input("Enter choice: ")

    # 1. Add Product
    if choice == "1":
        pid = input("Product ID: ")
        name = input("Product Name: ")
        price = float(input("Price: "))
        qty = int(input("Quantity: "))

        if qty >= 0:
            inventory[pid] = [name, price, qty]
            print("Product added successfully")
        else:
            print("Quantity cannot be negative")

    # 2. Update Stock
    elif choice == "2":
        pid = input("Product ID: ")

        if pid in inventory:
            add_qty = int(input("Enter quantity to add: "))
            if add_qty >= 0:
                inventory[pid][2] += add_qty
                print("Stock updated")
            else:
                print("Invalid quantity")
        else:
            print("Product not found")

    # 3. Purchase Products
    elif choice == "3":
        cart = []
        subtotal = 0

        while True:
            pid = input("Enter Product ID (or 'done'): ")
            if pid == "done":
                break

            if pid in inventory:
                qty = int(input("Enter quantity: "))

                if qty > 0 and qty <= inventory[pid][2]:
                    cost = qty * inventory[pid][1]
                    cart.append([inventory[pid][0], qty, cost])
                    inventory[pid][2] -= qty
                    subtotal += cost
                else:
                    print("Insufficient stock or invalid quantity")
            else:
                print("Product not found")

        # Billing
        tax = subtotal * 0.05

        discount = 0
        if subtotal >= 1000:
            discount = subtotal * 0.20
        elif subtotal >= 500:
            discount = subtotal * 0.10

        total = subtotal + tax - discount

        print("\n--- BILL ---")
        for item in cart:
            print(item[0], "-", item[1], "items - ₹", item[2])

        print("Subtotal: ₹", subtotal)
        print("Tax (5%): ₹", tax)
        print("Discount: ₹", discount)
        print("Total Payable: ₹", total)

    # 4. View Inventory
    elif choice == "4":
        print("\n--- INVENTORY ---")
        for pid in inventory:
            print(pid, inventory[pid][0],
                  "Price:", inventory[pid][1],
                  "Stock:", inventory[pid][2])

            if inventory[pid][2] < LOW_STOCK_LIMIT:
                print("⚠ Low Stock Alert")

    # 5. Exit
    elif choice == "5":
        print("Exiting system")
        break

    else:
        print("Invalid choice")
```
